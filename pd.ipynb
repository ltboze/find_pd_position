```python
#!/usr/bin/python
# -*- coding: UTF-8 -*-
import re
import requests
```

```python
def seek_pd(page=1,category='univ',date_limit='2019-01-01',keywords=[u'博士后']):
    if category == 'university':
        url = 'http://boshihoujob.com/gxbsh/list_144_'
    elif category == 'academy':
        url = 'http://boshihoujob.com/kybshzp/list_145_'
    elif category == 'enterprise':
        url = 'http://boshihoujob.com/qybsh/list_146_'
    else:
        pass
    
    for i in range(page):
        i += 1
        url += '{}.html'.format(i)
        url1 = 'http://boshihoujob.com'
        headers = {
            'User-Agent': 'Mozilla/5.0 (Macintosh; U; Intel Mac OS X 10_6_8; en-us) AppleWebKit/534.50 (KHTML, like Gecko) Version/5.1 Safari/534.50'}
        response = requests.get(url,headers=headers)
        response.encoding = 'gbk' #注意这里不用utf-8解码
        html = response.text
        ul = re.findall(r'<ul class="lst">(.*?)</ul>',html,re.S)[0]
        li = re.findall(r'<li>(.*)</li>',ul,re.S)[0]
        date = re.findall(r'<span>(.*?)</span>',li,re.S)
        tmp = re.findall(r'<h3><a href="(.*?)" class="title" target="_blank">(.*?)</a><\/h3>',li,re.S)

        for d,t in zip(date,tmp):
            if d >= date_limit:
                for keyword in keywords:
                    if keyword in t[1]:
                        if 'font' in t[1]:
                            print t[1][t[1].find("'>")+2:t[1].find("</")]
                            print url1+t[0]
                            print u'发布日期为：{}'.format(d)+'\n'
                            break
                        else:
                            print t[1]
                            print url1+t[0]
                            print u'发布日期为：{}'.format(d)+'\n'
                            break
                    else:
                        pass
```

```python
category = 'academy'
keywords = [u'中国科学院']
date_limit = '2019-03-01'
if __name__ == '__main__':
    %time seek_pd(page=1,category=category,date_limit=date_limit,keywords=keywords)
```
